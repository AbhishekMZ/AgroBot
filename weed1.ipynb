```python
from pyngrok import ngrok
import gradio as gr
from ultralytics import YOLO

model = YOLO("/kaggle/working/runs/detect/train2/weights/best.pt")

def detect_weed(image):
    results = model(image,conf=0.3)
    return results[0].plot()

interface = gr.Interface(fn=detect_weed, inputs="image", outputs="image", title="Weed Detector YOLOv8")
public_url = ngrok.connect(7860)
print(f"🔗 Gradio app running at: {public_url.public_url}")
interface.launch(share=True)
```
